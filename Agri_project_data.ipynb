```python
#Exploratory Data Analysis (EDA):

#Top 7 RICE PRODUCTION State Data(Bar_plot)
data_rice = df[["STATE_NAME","YEAR","RICE_PRODUCTION_1000_TONS"]]
```

```python
# Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart) 
data_wheat = df[["STATE_NAME","YEAR","WHEAT_PRODUCTION_1000_TONS"]]
```

```python
state_wheat = data_wheat.groupby("STATE_NAME")["WHEAT_PRODUCTION_1000_TONS"].sum().reset_index()
```

```python
top5_wheat_states = state_wheat.sort_values(by="WHEAT_PRODUCTION_1000_TONS", ascending=False).head(5)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top5_wheat_states, x="WHEAT_PRODUCTION_1000_TONS", y ="STATE_NAME" , color= "pink")
plt.title("Top 5 Wheat Producing States in India",fontsize=14)
plt.xlabel("Total Wheat Production(1000 tons)")
plt.ylabel("State")
plt.show()
```

```python
plt.figure(figsize=(5,5))

colors =  ["yellow", "blue", "green", "orange", "purple"]

plt.pie(
    top5_wheat_states["Percentage"],
    labels=top5_wheat_states["STATE_NAME"],
    autopct="%.1f%%",
    startangle=140,
    colors=colors,
    textprops={"color": "black"}   # 👈 makes text white
)

plt.title(
    "Percentage Share of Top 5 Wheat Producing States",
    fontsize=15,
    color="black"                  # 👈 title white
)

plt.gca().set_facecolor("black")   # optional: dark background
plt.show()
```

```python
# Impact of Area Cultivated on Production (Rice, Wheat, Maize)
data_area_production = df.groupby("YEAR")[[
    "RICE_PRODUCTION_1000_TONS", "RICE_AREA_1000_HA",
    "WHEAT_PRODUCTION_1000_TONS", "WHEAT_AREA_1000_HA",
    "MAIZE_PRODUCTION_1000_TONS", "MAIZE_AREA_1000_HA"
]].sum().reset_index()
```

```python
# Regression Line Plot
plt.figure(figsize=(12,6))
sns.regplot(x="RICE_AREA_1000_HA", y="RICE_PRODUCTION_1000_TONS", data=data_area_production, scatter_kws={'s':40}, label="Rice", color="pink")
sns.regplot(x="WHEAT_AREA_1000_HA", y="WHEAT_PRODUCTION_1000_TONS", data=data_area_production, scatter_kws={'s':40}, label="Wheat", color="green")
sns.regplot(x="MAIZE_AREA_1000_HA", y="MAIZE_PRODUCTION_1000_TONS", data=data_area_production, scatter_kws={'s':40}, label="Maize", color="seagreen")

plt.xlabel("Area Cultivated (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.title("Correlation between Area and Production")
plt.legend()
plt.show()
```

```python
# Rice vs. Wheat Yield Across States 
data_yield = df.groupby("STATE_NAME")[[
    "RICE_PRODUCTION_1000_TONS", "RICE_AREA_1000_HA",
    "WHEAT_PRODUCTION_1000_TONS", "WHEAT_AREA_1000_HA"
]].sum().reset_index()

data_yield["Rice_Yield"] = data_yield["RICE_PRODUCTION_1000_TONS"] / data_yield["RICE_AREA_1000_HA"]
data_yield["Wheat_Yield"] = data_yield["WHEAT_PRODUCTION_1000_TONS"] / data_yield["WHEAT_AREA_1000_HA"]

data_yield = data_yield[["STATE_NAME", "Rice_Yield", "Wheat_Yield"]]
```

```python
# Heatmap Plot
data_heat = data_yield.set_index("STATE_NAME")[["Rice_Yield","Wheat_Yield"]]

plt.figure(figsize=(8,10))
sns.heatmap(data_heat, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Rice vs Wheat Yield Across States (Heatmap)")
plt.show()
```
